```python
import numpy as np

def fit_linear_model(x, y):
    """
    Fit a linear model of the form y = ax + b to the given data x and y.

    Parameters:
    x (array-like): The x data.
    y (array-like): The y data.

    Returns:
    tuple: Coefficients (a, b) of the fitted linear model.
    """
    a, b = np.polyfit(x, y, 1)
    return a, b

def fit_linear_model_through_origin(x, y):
    """
    Fit a linear model of the form y = ax through the origin to the given data x and y.

    Parameters:
    x (array-like): The x data.
    y (array-like): The y data.

    Returns:
    float: Coefficient (a) of the fitted linear model.
    """
    A = np.vstack([x, np.ones(len(x))]).T
    a, _, _, _ = np.linalg.lstsq(A, y, rcond=None)
    return a[0]


# Example usage:
# x = np.linspace(-1,1,100)
# y = 2*x + 1 + np.random.randn(x.shape[0])  # y = 2x + 1 with some random noise

# a, b = fit_linear_model(x, y)
# print("Fitted model: y = {:.2f}x + {:.2f}".format(a, b))

a, b = fit_linear_model(fluctuations_true_trainset, fluctuations_trainset)
print("Fitted model: y = {:.2f}x + {:.2f}".format(a, b))

# a, b = fit_linear_model(fluctuations_trainset, fluctuations_true_trainset)
# print("Fitted model: y = {:.2f}x + {:.2f}".format(a, b))
```
